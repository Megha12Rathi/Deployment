#### Here is the meaning of each feature in the dataset:

Date: This column represents the date of a particular observation or record in the dataset. It likely indicates the date when the price and other property details were recorded.

Price: This column contains the price of a property or house. It represents the monetary value associated with the property.

Bedrooms: This column indicates the number of bedrooms in a house. It represents the count of individual rooms intended for sleeping purposes.

Bathrooms: This column represents the number of bathrooms in a house. It indicates the count of rooms containing toilet fixtures, such as a sink, toilet, and bathtub or shower.

Sqft_living: This column denotes the total living area or square footage of the house. It represents the interior space measured in square feet that includes living rooms, bedrooms, kitchen, etc.

Sqft_lot: This column indicates the total land area or square footage of the lot on which the house is built. It represents the size of the property's land in square feet.

Floors: This column represents the number of floors or levels in a house. It indicates the count of horizontal divisions in the structure where people can live or walk.

Waterfront: This column is a binary variable (0 or 1) indicating whether a property has a waterfront view (1) or not (0). It specifies whether the property is located adjacent to a body of water, such as a river, lake, or ocean.

View: This column represents the level of view the property has. It could be an ordinal variable indicating different levels of desirability or attractiveness of the view from the property.

Condition: This column represents the overall condition of the house. It is an ordinal variable that may indicate the quality, maintenance, or state of repair of the property.

Sqft_above: This column represents the square footage of the house above the ground level. It indicates the interior space measured in square feet for levels or floors that are above the ground.

Sqft_basement: This column denotes the square footage of the house's basement. It represents the interior space measured in square feet that is located below the ground level.

Yr_built: This column represents the year when the house was originally built or constructed. It indicates the age of the property in terms of years.

Yr_renovated: This column denotes the year when the house was last renovated. It indicates the most recent year when significant remodeling or renovation work was done on the property.

Street: This column likely represents the name or address of the street where the property is located.

City: This column denotes the name of the city where the property is situated.

Statezip: This column represents the state and zip code of the property. It provides information about the geographic location of the house.

Country: This column represents the country where the property is located. It specifies the nation to which the property belongs.

```python
columns_of_interest = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors']
```

```python
lower_quantile = 0.05  # Adjust the lower quantile threshold as needed
upper_quantile = 0.95  # Adjust the upper quantile threshold as needed

for column in columns_of_interest:
    lower_bound = df[column].quantile(lower_quantile)
    upper_bound = df[column].quantile(upper_quantile)
    df[column] = df[column].clip(lower_bound, upper_bound)
```

```python
df[columns_of_interest].boxplot()

# Set the title and labels
plt.title('Box Plot of Housing Features (After Handling Outliers)')
plt.xlabel('Features')
plt.ylabel('Values')
plt.show()
```

```python
categorical = df.dtypes[df.dtypes=='object'].index
print(categorical)

df[categorical].describe()
```

```python
from sklearn.preprocessing import LabelEncoder

categorical_columns = ['street', 'city', 'statezip', 'country']

label_encoder = LabelEncoder()
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])
```

```python
df.head()
```

```python
# Convert the "date" column to a pandas DateTime object
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Extract temporal features
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day_of_week'] = df['date'].dt.dayofweek
df['season'] = df['date'].dt.quarter
```

```python
# Total square footage
df['total_sqft'] = df['sqft_living'] + df['sqft_lot']

# Total number of bathrooms
df['cost_persqft'] = df['price']/ df['total_sqft']

# Age of the property
df['age'] = pd.Timestamp.now().year - df['yr_built']
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Select numerical columns for scaling
numerical_cols = ['price','bathrooms','bedrooms','sqft_living','sqft_lot','floors','waterfront','view','condition','city','statezip','country','month','day_of_week','season','total_sqft','cost_persqft','age']
# Scale numerical features
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
from sklearn.model_selection import train_test_split

# Select the relevant features for modeling
relevant_features = ['bedrooms','bathrooms' ,'floors','sqft_living', 'sqft_lot','waterfront', 'view', 'condition',
                     'city', 'statezip', 'country', 'month', 'day_of_week', 'season', 'total_sqft', 'cost_persqft', 'age']

X = df[relevant_features]
y = df['price']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score
```

```python
xgb_model = xgb.XGBRegressor(
    n_estimators=400,  # number of trees
    learning_rate=0.1,  # controls the step size shrinkage
    max_depth=4,  # maximum depth of each tree
    subsample=1.0,  # subsample ratio of the training instances
    colsample_bytree=0.8,  # subsample ratio of columns when constructing each tree
    random_state=42
)
```

```python
xgb_model.fit(X_train, y_train)
```

```python
y_pred = xgb_model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
